# Création des dataframes

```python
import os
import pandas as pd

def read_csv_files_in_directory(directory_path, encoding='utf-8'):
    csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]
    dataframes = {}
    
    for file in csv_files:
        file_path = os.path.join(directory_path, file)
        df_name = f"df_{os.path.splitext(file)[0]}"
        try:
            dataframes[df_name] = pd.read_csv(file_path, encoding=encoding)
            print(f"DataFrame '{df_name}' created with shape {dataframes[df_name].shape}")
        except UnicodeDecodeError:
            print(f"UnicodeDecodeError for file {file}. Trying 'ISO-8859-1' encoding...")
            dataframes[df_name] = pd.read_csv(file_path, encoding='ISO-8859-1')
            print(f"DataFrame '{df_name}' created with shape {dataframes[df_name].shape} using 'ISO-8859-1' encoding.")
    
    return dataframes
```

### Explication des Statuts de Prêt dans la Colonne "Status of Credit Bureau Loan"
#### Valeurs possibles et leurs significations :

1. **C** : **Closed (Clos)**
   - Ce statut indique que le prêt a été clôturé, signifiant généralement que le prêt a été entièrement remboursé et que l'emprunteur n'a plus de dettes envers le prêteur pour ce prêt spécifique.

2. **X** : **Status Unknown (Statut Inconnu)**
   - Ce statut signifie que le statut du prêt pour ce mois est inconnu ou n'a pas été rapporté. Il pourrait y avoir des raisons administratives pour lesquelles le statut n'a pas été enregistré, ou cela pourrait être une erreur de reporting.

3. **0** : **No DPD (Pas de retard)**
   - Cette valeur signifie qu'il n'y a eu aucun retard de paiement durant le mois. L'emprunteur a respecté toutes les obligations de paiement pour ce mois.

4. **1** : **DPD 1-30 (Retard de 1 à 30 jours)**
   - Cette valeur indique que le paiement mensuel était en retard de 1 à 30 jours durant le mois. C'est un indicateur que l'emprunteur a commencé à avoir des difficultés à rembourser son prêt à temps, mais qu'il n'est pas encore en grave défaut de paiement.

5. **2** : **DPD 31-60 (Retard de 31 à 60 jours)**
   - Cette valeur indique que le paiement mensuel était en retard de 31 à 60 jours. C'est une situation plus préoccupante, car cela montre que l'emprunteur a manqué plusieurs paiements ou qu'il est en retard depuis plus d'un mois.

6. **3, 4, 5, etc.** : **DPD prolongé (Retard prolongé)**
   - Les valeurs augmentent pour indiquer des retards plus longs :
     - **3** : Retard de 61 à 90 jours.
     - **4** : Retard de 91 à 120 jours.
     - **5** : Retard de 120+ jours, prêt vendu ou radié.
   - Un retard de 120 jours ou plus (valeur 5) est extrêmement grave et indique que le prêt pourrait avoir été vendu à une agence de recouvrement ou radié comme une perte par la banque.

#### Contexte :

Ces types de données sont couramment utilisées par les bureaux de crédit pour évaluer la solvabilité d'un emprunteur. Un historique de paiements réguliers (indiqué par beaucoup de `0`) est un bon signe, tandis que des DPD prolongés (par exemple, `2`, `3`, `4`, `5`) signalent des risques plus élevés et peuvent affecter négativement la cote de crédit d'un individu.
